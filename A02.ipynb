```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Boston Housing Dataset 1978.csv')
```

```python
df.describe()
```

```python
for i in df.columns:
    x = df[i]
    x_2 = (x-x.min())
    df[i] = x_2/x_2.max()
```

```python
df.describe()
```

```python
for i in df.columns:
    a = df[i]
    a_2 = (a - a.mean()) 
    std = a_2.std()
    df[i] = a_2/std

    
```

```python
import matplotlib.pyplot as plt
```

```python
df.hist(figsize=(15, 10), bins=30)
plt.tight_layout()
plt.show()
```
