```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
df = pd.read_csv("CLASE 6.csv")
```

```python
df.head(8)
```

```python
for i in df["Tamano"]:
    if i == "grande":
        df.loc[df["Tamano"] == "grande", "Tamano"] = 1
    else:
        df.loc[df["Tamano"] == "chica", "Tamano"] = 0


for i in df["hielo"]:
    if i == "con":
        df.loc[df["hielo"] == "con", "hielo"] = 1
    else:
        df.loc[df["hielo"] == "sin", "hielo"] = 0

df["Kahlua"] = df["sabor"].apply(lambda x: 1 if x == "Kahlua" else 0)
df["Horchata"] = df["sabor"].apply(lambda x: 1 if x == "Horchata" else 0)
   
    
```

```python
df.head(8)
```

```python
X = df[["Tamano", "hielo", "Kahlua", "Horchata"]]
y = df["score"]
lr = LinearRegression()
lr.fit(X, y)
print(f"El valor de Beta 0 es : {lr.intercept_}")
print(f"Los valores de Beta 1, 2 y 3 son : {lr.coef_}")
```

```python
for i in df["Tamano"]:
    if i == "grande":
        df.loc[df["Tamano"] == "grande", "Tamano"] = -1
    else:
        df.loc[df["Tamano"] == "chica", "Tamano"] = 1


for i in df["hielo"]:
    if i == "con":
        df.loc[df["hielo"] == "con", "hielo"] = 1
    else:
        df.loc[df["hielo"] == "sin", "hielo"] = -1

df["Kahlua"] = df["sabor"].apply(lambda x: 1 if x == "Kahlua" else 0)
df["Horchata"] = df["sabor"].apply(lambda x: 1 if x == "Horchata" else 0)
```

```python
X = df[["Tamano", "hielo", "Kahlua", "Horchata"]]
y = df["score"]
lr = LinearRegression()
lr.fit(X, y)
print(f"El valor de Beta 0 es : {lr.intercept_}")
print(f"Los valores de Beta 1, 2 y 3 son : {lr.coef_}")
```
