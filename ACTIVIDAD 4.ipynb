```python
import pandas as pd
from sklearn.gaussian_process import GaussianProcessRegressor
import numpy as np
```

```python
x = np.array([[10], [2], [3], [7], [12]]).reshape(-1, 1)
y = np.array([2847, -4017,-1255, 1773, 563])

x_pred = np.linspace(2,15,10001).reshape(-1, 1)
```

```python
modelo = GaussianProcessRegressor(random_state=42)
modelo.fit(x, y)
```

```python
import matplotlib.pyplot as plt

y_pred, sigma = modelo.predict(x_pred, return_std=True)

plt.figure(figsize=(10, 6))
plt.plot(x_pred, y_pred, label='Prediction')
plt.scatter(x, y, color='red', label='Training data')
plt.fill_between(x_pred.flatten(), y_pred - sigma, y_pred + sigma, alpha=0.3, label='Confidence interval')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gaussian Process Regression')
plt.legend()
plt.show()
```
