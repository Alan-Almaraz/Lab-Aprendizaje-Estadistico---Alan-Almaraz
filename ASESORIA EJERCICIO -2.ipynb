1.- Dos equipos juegan una serie de juegos (4 de 7, el que gane 4 juegos gana). Cada equipo tiene un 50% de probabilidad de ganar cada juego (no empates). ¿Cuál es la probabilidad de que la serie se vaya a 7 juegos? (5/16)

```python
import numpy as np
```

```python
def ss():
    v_a = 0
    v_b = 0
    total_juegos = 0

    while v_a < 4 and v_b < 4:
        total_juegos += 1
        resultado = np.random.randint(0, 2) 
        
        if resultado == 0:
            v_a += 1
        else:
            v_b += 1
    
    return total_juegos

num_simulaciones = 1000000
series_de_7 = 0

for i in range(num_simulaciones):
    if ss() == 7:
        series_de_7 += 1

probabilidad = series_de_7 / num_simulaciones
print(f"Probabilidad de llegar a 7 juegos: {probabilidad:.4f}")
```

2.- Una en mil personas tienen una enfermedad H, y la prueba de la enfermedad sale correcta 98% de las veces cuando la persona está enferma. Por otro lado, si la persona no está enferma la prueba tiene un error del 1%. Si alguien se hace una prueba y sale positiva, ¿cuál es la probabilidad de que tengan la enfermedad? (98/1097)

```python
n_p = 1000000
enfermos_reales = 0
positivos_totales = 0
enfermos_positivos = 0

for i in range(n_p):
    enfermo = np.random.rand() < 0.001  

    if enfermo:
        enfermos_reales += 1

        if np.random.rand() < 0.98:
            enfermos_positivos += 1
            positivos_totales += 1
    else:
        if np.random.rand() < 0.01:
            positivos_totales += 1

p = enfermos_positivos / positivos_totales

print(f"Probabilidad: {p:4f}")
```

3.- Tres amigos diferentes te dijeron que está lloviendo, y cada persona tiene 1/3 de probabilidad de mentir. Si el pronóstico del clima dijo que la probabilidad de lluvia era 25%, ¿cuál es la probabilidad de que sí esté lloviendo? (8/11)

```python
n_p = 1000000
llueve_realmente = 0
los_tres_dicen_lluvia = 0
llueve_y_dicen_lluvia = 0

for i in range(n_p):
    llueve = np.random.rand() < 0.25
    
    if llueve:
        a1_v = np.random.rand() < (2/3)
        a2_v = np.random.rand() < (2/3)
        a3_v = np.random.rand() < (2/3)
        
        if a1_v and a2_v and a3_v:
            los_tres_dicen_lluvia += 1
            llueve_y_dicen_lluvia += 1
    else:
        a1_m = np.random.rand() < (1/3)
        a2_m = np.random.rand() < (1/3)
        a3_m = np.random.rand() < (1/3)
        
        if a1_m and a2_m and a3_m:
            los_tres_dicen_lluvia += 1

p = llueve_y_dicen_lluvia / los_tres_dicen_lluvia

print(f"Probabilidad simulada: {p:.6f}")
```
