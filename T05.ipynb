### Carga de archivos y entendimiento de base de datos

```python
# Importamos librerías necesarias
import pandas as pd
from sklearn.linear_model import Ridge
import numpy as np
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#cargamos el dataset
df = pd.read_excel('Motor Trend Car Road Tests.xlsx')
```

```python
# Vemos información necesaria del dataset
df.info()
```

```python
#Nos familizarizamos con el dataset
df.head()
```

### Realiza una regresión tomando 'mpg' como salida y eliminando la columna 'model'. Considera todos los demás factores como numéricos/ordinales.

```python
# Nos pidieron usar el 40% de los datos para entrenar el modelo, en total son 32 datos, por lo que el 40% es igual a 13 datos
# Se seleccionan 13 datos de forma aleatoria para entrenar el modelo
np.random
m = df.sample(n=13, random_state=0)
print(m)

```

```python
# Primero antes de proceder escalamos los datos de entrenamiento
min_max_scaler = StandardScaler()
DE = min_max_scaler.fit_transform(m[["qsec", "cyl", "disp", "hp", "drat", "wt", "vs", "am", "gear", "carb"]])
sdf = pd.DataFrame(DE, columns=["qsec", "cyl", "disp", "hp", "drat", "wt", "vs", "am", "gear", "carb"])
sdf["mpg"] = m["mpg"].values

# Aquí mostramos nuestra escalamiento de los datos de entrenamiento
print(sdf.head())
```

```python
# Hacemos ya la regresión lineal con los datos escalados
X_train = sdf[["qsec", "cyl", "disp", "hp", "drat", "wt", "vs", "am", "gear", "carb"]]
y_train = sdf["mpg"]
lr = LinearRegression()
lr.fit(X_train, y_train)
r_2_entrenamiento = lr.score(X_train, y_train)
print("R^2:", r_2_entrenamiento)
print("Coeficientes:", lr.coef_)
print("Intercepto:", lr.intercept_)
```

```python
# Ya teniendo eso agarramos el 60% sobrante del dataset para probar el modelo
sobrantes = df.drop(m.index)

print(sobrantes)
```

```python
#   Escalamos los datos sobrantes
min_max_scaler = StandardScaler()
DE2 = min_max_scaler.fit_transform(sobrantes[["qsec", "cyl", "disp", "hp", "drat", "wt", "vs", "am", "gear", "carb"]])
sdf2 = pd.DataFrame(DE2, columns=["qsec", "cyl", "disp", "hp", "drat", "wt", "vs", "am", "gear", "carb"])
sdf2["mpg"] = sobrantes["mpg"].values
```

```python
# Predicciones con los datos sobrantes
x_pred = sdf2[["qsec", "cyl", "disp", "hp", "drat", "wt", "vs", "am", "gear", "carb"]]
y_pred = sdf2["mpg"]
modelo_entrenado = lr.predict(x_pred)
print("Predicciones:", modelo_entrenado)
print("Valores reales:", y_pred.values)
```

### Comparamos R2

```python
r2 = lr.score(x_pred, y_pred)
print("R2 del modelo en los datos sobrantes:", r2)
print("R2 del modelo en los datos de entrenamiento:", r_2_entrenamiento)
```

### Comparamos los Ridge

```python
# Ponemos los lambdas en una lista para probarlos todos al mismo tiempo con un ciclo for y no estar moviendo el codigo cada vez
lambdas = [0.1, 1.0, 10.0, 100.0]

for i in lambdas:
    ridge = Ridge(alpha=i)
    ridge.fit(X_train, y_train)
    r2_train = ridge.score(X_train, y_train)
    r2_test = ridge.score(x_pred, y_pred)
    print(f"Lambda: {i} | R2 Entrenamiento: {r2_train} | R2 Prueba: {r2_test}")
```

 A medida que el valor de lambda aumentaba, observamos que el R^2 de entrenamiento tendió a disminuir. Esto sucede porque la penalización L2 "obliga" a los coeficientes a ser más pequeños, impidiendo que el modelo se ajuste de forma perfecta a los datos de entrenamiento.
